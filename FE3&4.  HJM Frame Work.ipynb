## HJM Framework
 - class of models derived by directly modlelling the instantaneous forward rates
 - it provides explicit realtionship beween volatility of instantaeous forward rates and arbitrate free drift
 - it is a non markovian framework. only some models in this framework has closed for m solution

### Equilibrium(enogenous) models and non-arbitrage(exogenous) models
#### Endogenus models
- start with the assumption on the economic variables and derive a perocess for the short <br> rate
- first step is to calibrate to some set of chosen instruments and then obtain the yield<br> curve
##### examples     
vasicek model<br>
$dr(t) = \lambda(\theta - r(t) )dt + \eta dW(t) \label{eq:vasicek}\tag{3.1} $ 
       
       
Cox, Ingersoll, Ross model<br>        
$dr(t) = \lambda(\theta - r(t) )dt + \eta \sqrt{r(t)} dW(t) \label{eq:CIR}\tag{3.2}$

#### exogenous model 
- takes yield curve as input and then give the yield curve back . this is independento f the parameters
- the parmeters like mean revesion and  long term mean can be calibrated to different instruments
   

### Zero coupon bonds and forward rates relation
$P(t,t,T) = P(t,t,s) \times P(t,s,T)$ <br><br>
$\Rightarrow P(t,s,T) = \frac{P(t,t,T)}{P(t,t,s)} \label{eq:forward_bond}\tag{3.3}$ <br><br>

the forward rate from time t to T is represent as f(t,T) <br>
so $P(t,t,T) = e^{-f(t,s,T)} = e^{-r(t,s,T)\times (T-s)} = \frac{P(t,t,T)}{P(t,t,s)}$ <br><br>

taking log on both sides we get<br>
$e^{-r(t,s,T) } = -log \frac{P(t,t,T) - P(t,t,s)}{(T-s)}$ <br>

if T-s $\rightarrow $ 0, then we arrive at the definition of instantaneous forward rates

$$f(s,T) = e^{-r(t,s,T) } = -\frac{\partial}{\partial T} log P(t,T) \label{eq:inst_forward}\tag{3.4}$$ <br>





### Money savings account
Money saving account is important as it is taken as the default numeraire in the pricing of assets in the risk nuetral world. 

The money savings account is defined as 

$$ M(t) = exp \left(\int^t_0 r(s)ds \right) \equiv exp \left(\int^t_0 f(s,s)ds \right) \label{eq:money_savings_act}\tag{3.5}$$

### Instantaneous forward rate dynamics
$$df(t) = \mu(t,T)dt + \sigma(t,T) dW^\mathbb{P}(t) \label{eq:Inst_forw_rate_dynamics}\tag{3.6}$$
Note that the dynamics is given in real world measure

The three main bulding block for HJM framework are

- zero coupon bond relation to forward rates
- money savings account
- instataneous forward rate dynamics

The next step is to change the measure to $\mathbb{Q}$ such that the instantaneous forward rate dynamics is a martingale under this measure

### Arbitrage free conditions under the HJM model
#### Zero coupon bond and money saving account relation
Under the fundamental theorem of asset pricing ,  the value of any instrument under a chosen numeraire is a martingale under the probability measure induced by that numeraire. Note that the numeriare chosen has to be a tradable asset or a linear combination of tradable assets

$$ \frac{P(t,T)}{M(t,T)} =  \mathbb{E}^\mathbb{Q}\left[ \frac{P(T,T)}{M(T,T)}\mid \mathbb{F}(t) \right] \label{eq: Bond_Value1}\tag{3.7}$$ <br> <br>
$ P(t,T)= \mathbb{E}^\mathbb{Q}\left[ exp\left( - \int^T_t(r(s)ds \right)\mid \mathbb{F}(t) \right] \label{eq: Bond_Value2}\tag{3.8}$  <br>

as $P(T,T)$ and $M(t,T) $ equals to 1



The final HJM arbitrage free model is 
$$\boxed{df(t,T) = \alpha ^\mathbb{Q}(t,T) dt + \sigma(t,T) dW^\mathbb{Q} (t) }\label{eq: HJM_Model}\tag{3.9} $$
the  no arbirage drift condition is given by 

$$\color{blue}{ \alpha ^\mathbb{Q}(t,T) = \sigma(t,T) \int_t^T \sigma (t,s) ds }\label{eq:HJM_Drift} \tag{3.10}$$

- Now by using the relation $f(t,t) = r(t)$ and integrating the above SDEs, we obtain the short rate dynamics under HJM framework
$$f(t,T) = f(0,T) + \int_0^t \alpha^\mathbb{Q}(s,T) ds + \int_0^t \sigma(s,T) dW^\mathbb{Q}(s)$$

now as per the short rate definition $T =t$ and hence the above equation becomes

$$f(t,t) = r(t) = f(0,t) + \int_0^t \alpha^\mathbb{Q}(s,t) ds + \int_0^t \sigma(s,t) dW^\mathbb{Q}(s) $$


we need the the derivative of the short rate $r(t)$ for finding its dynamics. This is obtained by appying the Lebniz rule . We get

$$ \boxed{dr(t) = \left[ \frac{\partial}{\partial t}f(0,t) +  \alpha^\mathbb{Q}(t,t) ds + \int_0^t \frac{\partial}{\partial t} \alpha ^\mathbb{Q}(s,t)ds +  \int_0^t \frac{\partial}{\partial t}\sigma(s,t) dW^\mathbb{Q}(s) \right]\times dt + \sigma(t,t) dW^\mathbb{Q}(t) }\label{eq:Short_Rate_Dynamics} \tag{3.11}$$

# Ho and Lee model
- by specifing the volatility , we can have different models from the HJM frame work. 
- for HO Lee model , we define volatility as
$ \sigma(t,T) = \sigma \label{eq:Ho Lee vol}  \tag{3.12}$ <br> 
ie the volatility is constant <br>

Now substituting the above defined volatility to the HJM arbitrage free condition for the drift we get

$\alpha^\mathbb{Q}(t,T) = \sigma \; \int_t^T \sigma ds = \sigma^2(T-t)
\label{eq:Ho Lee drift} \tag{3.13}
$

The above equations can be substituted in the short rate dynamics equation $\{eq:Short_Rate_Dynamics}$ 

After simplification we get the dynamics of the interest rate under HO-Lee Model as 
$$ dr(t) = \left(\frac{\partial}{\partial t} f(0,t) + \sigma^2 t \right) dt + \sigma dW^{\mathbb{Q}} \label{eq:Ho Lee Model 1} \tag{3.14}$$

By setting $ \theta(t) = \frac{\partial}{\partial t} f(0,t) + \sigma^2 t $, we arrive at 

 $$ \boxed{\color{blue} {dr(t) = \theta(t) dt + \sigma dW^{\mathbb{Q}} }}\label{eq:Ho Lee Model 2} \tag{3.15}$$



As the Ho Lee Model fall in the affine class of processes, we can write it as

```python
#%%
"""
Created on July 12 2021
Ho-Lee Model, Simulation of the Model + Computation of ZCBs, P(0,t)

This code is purely educational and comes from "Financial Engineering" course by L.A. Grzelak
The course is based on the book “Mathematical Modeling and Computation
in Finance: With Exercises and Python and MATLAB Computer Codes”,
by C.W. Oosterlee and L.A. Grzelak, World Scientific Publishing Europe Ltd, 2019.
@author: Lech A. Grzelak
"""

import numpy as np
import matplotlib.pyplot as plt

def f0T(t,P0T):
    # time-step needed for differentiation
    dt = 0.01    
    expr = - (np.log(P0T(t+dt))-np.log(P0T(t-dt)))/(2*dt)
    return expr

def GeneratePathsHoLeeEuler(NoOfPaths,NoOfSteps,T,P0T,sigma):    
    
    # Initial interest rate is a forward rate at time t->0
    r0 = f0T(0.01,P0T)
    theta = lambda t: (f0T(t+dt,P0T)-f0T(t-dt,P0T))/(2.0*dt) + sigma**2.0*t 
     
    Z = np.random.normal(0.0,1.0,[NoOfPaths,NoOfSteps])
    W = np.zeros([NoOfPaths, NoOfSteps+1])
    R = np.zeros([NoOfPaths, NoOfSteps+1])
    M = np.zeros([NoOfPaths, NoOfSteps+1])
    M[:,0]= 1.0
    R[:,0]=r0
    time = np.zeros([NoOfSteps+1])
        
    dt = T / float(NoOfSteps)
    for i in range(0,NoOfSteps):
        # making sure that samples from normal have mean 0 and variance 1
        if NoOfPaths > 1:
            Z[:,i] = (Z[:,i] - np.mean(Z[:,i])) / np.std(Z[:,i])
        W[:,i+1] = W[:,i] + np.power(dt, 0.5)*Z[:,i]
        R[:,i+1] = R[:,i] + theta(time[i]) * dt + sigma* (W[:,i+1]-W[:,i])
        M[:,i+1] = M[:,i] * np.exp((R[:,i+1]+R[:,i])*0.5*dt)
        time[i+1] = time[i] +dt
        
    # Outputs
    paths = {"time":time,"R":R,"M":M}
    return paths

def mainCalculation():
    NoOfPaths = 25000
    NoOfSteps = 500
       
    sigma = 0.007
        
    # We define a ZCB curve (obtained from the market)
    P0T = lambda T: np.exp(-0.1*T)
       
    # In this experiment we compare ZCB from the Market and Monte Carlo
    "Monte Carlo part"   
    T = 40
    paths= GeneratePathsHoLeeEuler(NoOfPaths,NoOfSteps,T,P0T,sigma)
    M = paths["M"]
    ti = paths["time"]
        
    # Here we compare the price of an option on a ZCB from Monte Carlo and Analytical expression    
    P_t = np.zeros([NoOfSteps+1])
    for i in range(0,NoOfSteps+1):
        P_t[i] = np.mean(1.0/M[:,i])
   
    plt.figure(1)
    plt.grid()
    plt.xlabel('T')
    plt.ylabel('P(0,T)')
    plt.plot(ti,P0T(ti))
    plt.plot(ti,P_t,'--r')
    plt.legend(['P(0,t) market','P(0,t) Monte Carlo'])
    plt.title('ZCBs from Ho-Lee Model')
    
mainCalculation()
```

### Hull-White Model
- Hull white model is also another special case of HW model
- It is mean reverting model
- the volatility for HW model is defined as 
$\sigma(t,T) = \sigma \times e^{-\lambda(T-t)} \label{eq: HW Model Vol}\tag{3.17}$ where $\lambda > 0$


Subsituting the above in the volatillity expression in the HJM short rate dynamics derived earlier in equation $\ref{eq:Short_Rate_Dynamics} $, and applying algebraic simplifications (to be detailed later) we finally get 

$dr(t) = \lambda(\theta(t) -r(t) )dt + \sigma dW^{\mathbb{Q}}(t) \label{eq:HW model} \tag{3.18}$

where $\theta(t) $  is
$ \theta(t) = \frac{1}{\lambda} \frac{\partial}{\partial t} f(0,t) + f(0,t) \;+\; \frac{\sigma^2}{2\lambda^2}\left( 1- e^{-2 \lambda t}\right) \label{eq:HW model theta} \tag{3.19}$


The $\theta(t)$ function guarantees that the zero coupon bond obtained from integrting the short rates, then using its exponent and taking the expectation of that  is same as the original zero coupon bond input to the model

choosing initial value process for  $r(0)$ in the $\theta(t)$ expression is very important ( add details later)

$$ \color{red}{r(0) = f(0,0) \approx -\frac{\partial log\; P(0,\epsilon)}{\partial \epsilon}}$$ for $\epsilon \rightarrow 0$

- the lambda and sigma parameters are free parameters. These parameters ar calibrated to the type instruments which we are interesteed in pricing ( for example  swaptions, caps ets). In practice the sigma(volatility) parameter is calibrated very frequently and the lambda parameter is calibrated once a week/ month   
- also the model calibration shuld be done using the instruments which are used for hedging the traded instrument. this means that the instruments chosen for calibration would be the liquid vanilla instruments 
- as the input yield curve is fully recovered as the ouput of the model(independent of the lamda and sigma cosen) , the additional parameters lambda and theta can be used for calibration to specific instruments

```python
#%%
"""
Created on July 12 2021
Monte Carlo paths for the Hull-White model

This code is purely educational and comes from "Financial Engineering" course by L.A. Grzelak
The course is based on the book “Mathematical Modeling and Computation
in Finance: With Exercises and Python and MATLAB Computer Codes”,
by C.W. Oosterlee and L.A. Grzelak, World Scientific Publishing Europe Ltd, 2019.
@author: Lech A. Grzelak
"""
import numpy as np
import matplotlib.pyplot as plt

def GeneratePathsHWEuler(NoOfPaths,NoOfSteps,T,P0T, lambd, eta):    
    # time-step needed for differentiation
    dt = 0.0001    
    f0T = lambda t: - (np.log(P0T(t+dt))-np.log(P0T(t-dt)))/(2*dt)
    
    # Initial interest rate is a forward rate at time t->0
    r0 = f0T(0.00001)
    theta = lambda t: 1.0/lambd * (f0T(t+dt)-f0T(t-dt))/(2.0*dt) + f0T(t) + \
    eta*eta/(2.0*lambd*lambd)*(1.0-np.exp(-2.0*lambd*t))      
    
    Z = np.random.normal(0.0,1.0,[NoOfPaths,NoOfSteps])
    W = np.zeros([NoOfPaths, NoOfSteps+1])
    R = np.zeros([NoOfPaths, NoOfSteps+1])
    R[:,0]=r0
    time = np.zeros([NoOfSteps+1])
        
    dt = T / float(NoOfSteps)
    for i in range(0,NoOfSteps):
        # making sure that samples from normal have mean 0 and variance 1
        if NoOfPaths > 1:
            Z[:,i] = (Z[:,i] - np.mean(Z[:,i])) / np.std(Z[:,i])
        W[:,i+1] = W[:,i] + np.power(dt, 0.5)*Z[:,i]
        R[:,i+1] = R[:,i] + lambd*(theta(time[i]) - R[:,i]) * dt + eta* (W[:,i+1]-W[:,i])
        time[i+1] = time[i] +dt
        
    # Outputs
    paths = {"time":time,"R":R}
    return paths

def mainCalculation():
    NoOfPaths = 1
    NoOfSteps = 5000
    T         = 50.0
    lambd     = 0.5
    eta       = 0.01
    
    # We define a ZCB curve (obtained from the market)
    P0T = lambda T: np.exp(-0.05*T) 

    # Effect of mean reversion lambda
    plt.figure(1) 
    legend = []
    lambdVec = [-0.01, 0.2, 5.0]
    for lambdTemp in lambdVec:    
        np.random.seed(2)
        Paths = GeneratePathsHWEuler(NoOfPaths,NoOfSteps,T,P0T, lambdTemp, eta)
        legend.append('lambda={0}'.format(lambdTemp))
        timeGrid = Paths["time"]
        R = Paths["R"]
        plt.plot(timeGrid, np.transpose(R))   
    plt.grid()
    plt.xlabel("time")
    plt.ylabel("R(t)")
    plt.legend(legend)
        
    # Effect of the volatility
    plt.figure(2)    
    legend = []
    etaVec = [0.1, 0.2, 0.3]
    for etaTemp in etaVec:
        np.random.seed(2)
        Paths = GeneratePathsHWEuler(NoOfPaths,NoOfSteps,T,P0T, lambd, etaTemp)
        legend.append('eta={0}'.format(etaTemp))
        timeGrid = Paths["time"]
        R = Paths["R"]
        plt.plot(timeGrid, np.transpose(R))   
    plt.grid()
    plt.xlabel("time")
    plt.ylabel("R(t)")
    plt.legend(legend)
    
mainCalculation()
```

### Solution to Hull-WHite SDE
- to obtain solution to hull white sde, we select the following process and apply itos lemma to it
$y(t) = e^{\lambda t }r(t) \label{HW sol1} \tag{3.20}$ 

applying itos lemma to it we get, 

$dy(t) =\lambda y(t) dt + e^{\lambda t} dr(t) \label{HW sol2} \tag{3.21}$

the dynamics of HW model is given in the section above as
$dr(t) = \lambda(\theta(t) -r(t) )dt + \sigma dW^{\mathbb{Q}}(t) \label{eq:HW model copy} $
substituting the HW dynamics in the dynamics for process $y$, we get
$$dy(t) = \lambda y(t) dt + e^{\lambda t}\left[ \lambda (\theta(t) - r(t))dt + \eta dW_r(t) \right] $$

substituting the value of $y$
$$dy(t) = \color{red}{\lambda e^{\lambda t }r(t) dt} + \color{red}{e^{\lambda t}}\left[ \color{red}{\lambda} (\theta(t) - \color{red}{r(t)})\color{red}{dt} + \eta dW_r(t) \right] $$


Cancelling out like terms we get 

$$dy(t) = \lambda \theta(t) e^{\lambda t} dt + \eta e^{\lambda t} dW_r(t)$$

note that the right hand side doenot depend on the state vector y(t0 now and we can therefore determine the solution by integrating both sides of sde

$$\int _0^t dy(z) = \lambda \int^t_0 \theta(z) e^{\lambda z} dz + \eta\int _0 ^t e^{\lambda z } dW_r (z) $$

which gives us,

$$y(t) = y(0) + \lambda \int^t_0 \theta(z) e^{\lambda z} dz + \eta\int _0 ^t e^{\lambda z } dW_r (z) $$

apply the relationship between $y(t)$ and $r(t)$, we get
$$r(t) = \underbrace{e^{-\lambda t}r(0) + \lambda \int^t_0 \theta(z) e^{-\lambda(t- z)}dz}_{deterministic}  + \underbrace{\eta\int _0 ^t e^{-\lambda (t-z) } dW_r (z)}_{stochastic} \label{HW sol final} \tag{3.22}$$
<br><br>
The mean of the $r(t)$ is its determisnistic part $$ \mathbb{E}\left[ r(t) | \mathbb{F}(t_0)\right] \;=\; e^{-\lambda t}r(0) + \lambda \int^t_0 \theta(z) e^{-\lambda(t- z)} dz$$

<br> 
The general formula for variance is  
$varinace = E[X^2] - E[X]^2$

The variance for $r(t)$driven by the stochastic part only . It is given by $\color{red}{( add \;derivation  \; later)}$

$$ \mathbb{V}\left[ r(t) | \mathbb{F}(t_0)\right] = \frac{\eta^2}{2\lambda}(1-e^{-2\lambda t})$$


As the HW model belongs to affine class of processes , we can represent the ZCB as 
$$P(t,T) = \mathbb{E}^\mathbb{Q} \left[ e^{-\int_t^T r(z)dz  } |\mathbb{F}(t) \right] \; = \; e^{A(0,\tau) + B(0,\tau) r(t)}$$
<br>
The analytical solution for A and B are given by 

$\boxed{ \begin{equation}
A(0,\tau) =  -\frac{\eta^2}{4\lambda^3}(3+ e^{-2 \lambda \tau} -4e^{-\lambda \tau} -2\lambda \tau) + \lambda \int_0^\tau {\theta(T-z) \tilde{C}(u,z)dz} \\ 
B(0,\tau) = -\frac{1}{\lambda}(1- e^{-\lambda(T-t)})
\end{equation} } \label{HW Affine A & B} \tag{3.22}$
